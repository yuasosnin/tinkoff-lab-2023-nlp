```python
import matplotlib.pyplot as plt
import numpy as np

metric_data = {
    "sft": [0.5132103729248046, 5.681484701582379],
    "hinge": [2.8122852783203127, 5.197111852268359],
    "sigmoid": [2.8304215087890623, 5.093427006400522],
    "fowrard-kl": [2.821338623046875, 5.091937844807061],
    "alpha-kl-0.3": [2.59627685546875, 5.08950023832391],
    "alpha-kl-0.5": [2.72556201171875, 5.05890483763371],
    "alpha-kl-0.7": [2.741251708984375, 5.0682273563575055],
    "js": [2.8037750244140627, 5.109860521435615]
}
```

```python
data = np.array(list(metric_data.values()))
x = np.linspace(min(data.T[0]) - 0.1, max(data.T[0]) + 0.1, 100)

z = np.polyfit(*data.T, 2)
p = np.poly1d(z)
```

```python
fig, ax = plt.subplots()
for name, obs in metric_data.items():
    ax.scatter(*obs, label=name)
plt.plot(x, p(x),"r--")
plt.legend(loc='lower left')
plt.title("Reward vs. Entropy")
plt.xlabel("Reward")
plt.ylabel("Entropy")
plt.show()
```
